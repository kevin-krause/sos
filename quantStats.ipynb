### Quant Stats:
###### Biblioteca que vamos utilizar para vizualisar quant

###### https://pypi.org/project/QuantStats/

```python
import quantstats as qs
```

```python
# extend pandas functionality with metrics, etc.
qs.extend_pandas()

# fetch the daily returns for a stock
stock = qs.utils.download_returns('^BVSP')

# show sharpe ratio
qs.stats.sharpe(stock)

# or using extend_pandas() :)
stock.sharpe()
```

```python
# Vizualizar a performance
qs.plots.snapshot(stock, title='Stock Performance')

# can also be called via:
# stock.plot_snapshot(title='Stock Performance')
```

```python
#(benchmark can be a pandas Series or ticker)
qs.reports.full(stock, "PETR4.SA", )
```
